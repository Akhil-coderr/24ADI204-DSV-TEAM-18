# Introduction to Olympics

The modern Olympic Games or Olympics are leading international sporting events featuring summer and winter sports competitions in which thousands of athletes from around the world participate in a variety of competitions. The Olympic Games are considered the world’s foremost sports competition with more than 200 nations participating. The Olympic Games are held every four years, with the Summer and Winter Games alternating by occurring every four years but two years apart.

The evolution of the Olympic Movement during the 20th and 21st centuries has resulted in several changes to the Olympic Games. Some of these adjustments include the creation of the Winter Olympic Games for snow and ice sports, the Paralympic Games for athletes with a disability, the Youth Olympic Games for athletes aged 14 to 18, the five Continental games (Pan American, African, Asian, European, and Pacific), and the World Games for sports that are not contested in the Olympic Games. The Deaflympics and Special Olympics are also endorsed by the IOC. The IOC has had to adapt to a variety of economic, political, and technological advancements.

 As a result, the Olympics has shifted away from pure amateurism, as envisioned by Coubertin, to allowing participation of professional athletes. The growing importance of mass media created the issue of corporate sponsorship and commercialisation of the Games. 
 
 World wars led to the cancellation of the 1916, 1940, and 1944 Games. Large boycotts during the Cold War limited participation in the 1980 and 1984 Games. The latter, however, attracted 140 National Olympic Committees, which was a record at the time.

**A standard statistical method called the Boxplot method is used to find the "fences" of our data,any value outside these bounds is mathematically considered an outlier.**

```python
# IQR method
Q1 = df['Age'].quantile(0.25) # The median of the lower half of the data
Q3 = df['Age'].quantile(0.75) # The median of the upper half of the data
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR  # Lower Fence
upper_bound = Q3 + 1.5 * IQR  # Upper Fence

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Count outliers
outlier_count = ((df['Age'] < lower_bound) | (df['Age'] > upper_bound)).sum() # Any data point that is less than the Lower Fence or greater than the Upper Fence is an outlier.
print("Number of Age outliers:", outlier_count)
```

##  Age Outlier Handling – Olympic Dataset

###  Initial Statistical Analysis
To detect outliers in the **Age** column, the Interquartile Range (IQR) method was initially applied.

- **Lower fence (Q1 − 1.5 × IQR):** 13  
- **Upper fence (Q3 + 1.5 × IQR):** 37  

According to pure statistical rules, athlete ages outside the range **13–37** would be classified as outliers.

---

###  Issue with Purely Statistical Approach
While the IQR method is statistically valid, it does **not account for domain-specific realities** of the Olympic Games:

- The Olympics span **over 120 years**
- Certain sports allow:
  - **Very young athletes** (historically as young as 10)
  - **Older athletes**, especially in precision and low-impact sports

Using the IQR-based fences would incorrectly remove **legitimate Olympic medalists**.

---

###  Domain Investigation & Historical Evidence
Upon investigating historical Olympic records:

- The **youngest medal-winning athletes** was in history is Greek gymnast Dimitrios Loundras, who competed in the **1896 Athens Olympics at 10 years and 218 days old, winning a bronze in team parallel bars**
- The **oldest medalist** was around **72 years old** is Swedish shooter Oscar Swahn, who won a silver medal at the **1920 Antwerp Games at the age of 72 years and 281 days, also making him the oldest Olympian to compete.**

To safely include all legitimate cases, the **upper bound was extended to 73 years**.

---

###  Final Decision (Domain-Aware Clipping)
Instead of removing outliers, **winsorization (clipping)** was applied using historically valid limits:

- **Lower bound:** 10 years  
- **Upper bound:** 73 years  

This ensures:
- Preservation of historically valid athletes
- Prevention of extreme age values from skewing analysis
- A balance between statistical robustness and real-world correctness

---

###  Conclusion
Although the IQR method suggested an age range of **13–37**, domain knowledge and historical Olympic data justified extending the clipping range to **10–73**.  
This approach results in a cleaner, more realistic, and analytically sound dataset.


```python
# Capping Age outliers using IQR limits
df['Age'] = df['Age'].clip(lower=10, upper=73)
```


**IQR-based capping(winsorization) or ROBUST SCALING** replaces outliers using quartile boundaries. It is preferred because it effectively handles skewed, non-normal data, whereas Z-scores assume a normal distribution.

Unlike Z-scores, which rely on the mean (sensitive to outliers), IQR uses stable percentiles, making it far more robust for real-world analysis.

```python
# Recheck Age range after treatment
print("New Min Age:", df['Age'].min())
print("New Max Age:", df['Age'].max())

# Distribution check
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution After Outlier Handling")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Boxplot after treatment
plt.figure(figsize=(8, 3))
sns.boxplot(x=df['Age'])
plt.title("Boxplot of Age After Outlier Treatment")
plt.xlabel("Age")
plt.show()
```

Although ages above the IQR upper fence still appear as outliers in the boxplot, these values are not treated as errors. 

They represent statistically rare but historically valid Olympic athletes. 

Since boxplots identify outliers purely based on distribution spread, these points were retained after domain-aware clipping

## Outlier handling using Z-scores stratagey

```python
# Step 1: Basic range check
print("Original Height Range:")
print("Min Height:", df['Height'].min())
print("Max Height:", df['Height'].max())


# Step 2: Distribution check

plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Height'])
plt.title("Boxplot of Height Before Z-Score Outlier Handling")
plt.xlabel("Height")
plt.show()


# Step 3: Z-score using sklearn
scaler = StandardScaler()
height_z = scaler.fit_transform(df[['Height']])

# Step 4: Identify outliers
z_threshold = 3
outlier_count = (np.abs(height_z) > z_threshold).sum()

print("\nZ-Score Threshold:", z_threshold)
print("Number of Height Outliers:", outlier_count)


# Step 5: Cap outliers
height_mean = df['Height'].mean()
height_std = df['Height'].std()

upper_limit = height_mean + z_threshold * height_std
lower_limit = height_mean - z_threshold * height_std

df['Height'] = df['Height'].clip(lower=lower_limit, upper=upper_limit)


# Step 6: Validation (Recalculate Z-score after clipping)


print("\nHeight Range After Z-score Outlier Handling:")
print("Min Height:", df['Height'].min())
print("Max Height:", df['Height'].max())



# Step 7: Visual validation
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Height'])
plt.title("Boxplot of Height After Z-Score Outlier Handling")
plt.xlabel("Height")
plt.show()

```

**Z-score based capping did not fully eliminate extreme values due to its dependence on mean and standard deviation.**

**To achieve a more robust and distribution-independent treatment, the Robust Scaling method was applied instead.**


```python
# Step 1: Basic range check
print("Original Height Range:")
print("Min Height:", df['Height'].min())
print("Max Height:", df['Height'].max())


# # Boxplot before outlier handling
plt.figure(figsize=(8, 3))
sns.boxplot(x=df['Height'])
plt.title("Boxplot of Height Before IQR Outlier Handling")
plt.xlabel("Height")
plt.show()


# Step 3: IQR calculation
Q1 = df['Height'].quantile(0.25)
Q3 = df['Height'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


# Step 4: Identify outliers BEFORE handling
outliers_before = ((df['Height'] < lower_bound) | (df['Height'] > upper_bound)).sum()

print("\nIQR Lower Bound:", lower_bound)
print("IQR Upper Bound:", upper_bound)
print("Number of Height Outliers BEFORE handling:", outliers_before)


# Step 5: Cap outliers using clip()
df['Height'] = df['Height'].clip(lower=lower_bound, upper=upper_bound)


# Step 6: Identify outliers AFTER handling
outliers_after = ((df['Height'] < lower_bound) | (df['Height'] > upper_bound)).sum()


# Step 7: Validation
print("\nHeight Range After IQR Outlier Handling:")
print("Min Height:", df['Height'].min())
print("Max Height:", df['Height'].max())

print("\nNumber of Height Outliers AFTER handling:", outliers_after)


# Boxplot after handling
plt.figure(figsize=(8, 3))
sns.boxplot(x=df['Height'])
plt.title("Boxplot of Height After IQR Outlier Handling")
plt.xlabel("Height")
plt.show()

```

```python
# Step 1: Basic range check
print("Original Weight Range:")
print("Min Weight:", df['Weight'].min())
print("Max Weight:", df['Weight'].max())


# Step 2: Boxplot Before outlier handling
plt.figure(figsize=(8, 3))
sns.boxplot(x=df['Weight'])
plt.title("Boxplot of Weight Before IQR Outlier Handling")
plt.xlabel("Weight")
plt.show()

# Step 3: IQR calculation
Q1 = df['Weight'].quantile(0.25)
Q3 = df['Weight'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


# Step 4: Identify outliers BEFORE handling
outliers_before = ((df['Weight'] < lower_bound) | (df['Weight'] > upper_bound)).sum()

print("\nIQR Lower Bound:", lower_bound)
print("IQR Upper Bound:", upper_bound)
print("Number of Weight Outliers BEFORE handling:", outliers_before)


# Step 5: Cap outliers using clip()
df['Weight'] = df['Weight'].clip(lower=lower_bound, upper=upper_bound)


# Step 6: Identify outliers AFTER handling
outliers_after = ((df['Weight'] < lower_bound) | (df['Weight'] > upper_bound)).sum()


# Step 7: Validation
print("\nWeight Range After IQR Outlier Handling:")
print("Min Weight:", df['Weight'].min())
print("Max Weight:", df['Weight'].max())

print("\nNumber of Weight Outliers AFTER handling:", outliers_after)


# Boxplot after handling
plt.figure(figsize=(8, 3))
sns.boxplot(x=df['Weight'])
plt.title("Boxplot of Weight After IQR Outlier Handling")
plt.xlabel("Weight")
plt.show()


```

The IQR method was applied to detect extreme values in the Weight column.
Outliers were capped using statistical bounds to reduce skewness while preserving all records.
